```python
#Models
def predict_point_by_point(model, data):
    predicted = model.predict(data)
    predicted = np.array(predicted)
    predicted = np.reshape(predicted, (predicted.size,))
    return predicted

def plot_results(predicted_data, true_data):
    fig = plt.figure(facecolor='white')
    plt.plot(true_data, label='TrueData')
    plt.plot(predicted_data, label='Prediction')
    plt.legend()
    plt.show()
```

```python
from sklearn.utils import class_weight
class_weights = class_weight.compute_class_weight('balanced',
                                                 np.unique(y_train),
                                                 y_train)
class_weights
```

```python
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation 
from keras.utils import np_utils
from keras import optimizers


Y_train = np_utils.to_categorical(1*(y_train>=0), 2) 
Y_test = np_utils.to_categorical(1*(y_test>=0), 2) 

np.random.seed(100)

#Denses: 128+16+1?
model = Sequential()
model.add(Dense(512, input_shape=(len(X_train.iloc[0,:]),))) #kernel_initializer?
model.add(Activation('relu'))
model.add(Dropout(0.2)) # 
model.add(Dense(128)) 
model.add(Activation('relu')) 
model.add(Dropout(0.2)) 
#model.add(Dense(256)) #we proved by adding another layer
#model.add(Activation('relu'))
#model.add(Droput(0.2))
model.add(Dense(2))
model.add(Activation('softmax')) #sigmoide activation (imagen entre 0 y 1)

sgd = optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True) #low learning rate (typically 0.01 or 0.001)
model.compile(loss='categorical_crossentropy', optimizer=sgd,metrics=['accuracy',keras.metrics.Precision(), keras.metrics.Recall()])

#model.compile(loss='categorical_crossentropy', 
#              optimizer='adam',metrics=['accuracy',keras.metrics.Precision(), keras.metrics.Recall()]) #lsgd 
#include learning rate¡
#optimizer
model.fit(X_train, Y_train,batch_size=128, epochs=10, class_weight=class_weights,
                    verbose=0, validation_data=(X_test, Y_test),shuffle=True)#shuffling to avoid reach the function constant
#to 1. Since we get consecutively many +1 or -1, it may solve 

          #muchos ejempplos a la vez. epochs numero de vueltas al cto de entrenamiento entero. Early stopping
          #cuando el cto de validacion deje de mejorar, deja de entrenar. 

score = model.evaluate(X_test, Y_test, verbose=0)
```
